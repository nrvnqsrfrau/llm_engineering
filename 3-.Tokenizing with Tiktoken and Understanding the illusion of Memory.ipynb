```python
# Tokenize with code
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")
tokens = encoding.encode("Hi my name is Reza and I like Mie Ayam")
```

```python
tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
```

<h2>The Illusion of "Memory"</h2>

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
openai = OpenAI()

load_dotenv(override=True)
# api_key = os.getenv("OPENAPI_API_KEY")
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key="ollama")
```

```python
messages = [
    {"role":"system", "content":"You are a helpful assistant"},
    {"role":"user", "content":"Hi! I'm Reza"}
]
```

```python
response = ollama.chat.completions.create(model="llama3.2",messages=messages)
response.choices[0].message.content
```

Follow Up Question

```python
messages = [
    {"role":"system","content":"You are a helpful assistant"},
    {"role":"user","content":"What's my name ?"}
]
```

```python
response = ollama.chat.completions.create(model="llama3.2",messages=messages)
response.choices[0].message.content
```

Every call to a LLM is completely STATELESS. It's a totally new call every single time. It's time to devise techniques to give the impression that the LLM has a "Memory"

```python
messages = [
    {"role":"system","content":"You are a helpful assistant"},
    {"role":"user","content":"Hi! I'm Reza"},
    {"role":"assistant","content":"Hi Reza! How can i assist you today ?"},
    {"role":"user","content":"What's my name ?"}
]

response = ollama.chat.completions.create(model="llama3.2",messages=messages)
response.choices[0].message.content
```
